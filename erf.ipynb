```python
# error function versus coulomb potential
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# Constants
Z = 1  # Nuclear charge
e2 = 1  # In atomic units, e^2 = 1
Rc_values = [0.2,1.5, 1.75, 2.0, 2.5]  # Different Rc values for comparison

# Define radial coordinate grid
r = np.linspace(0.01, 5.0, 1000)  # Avoid r=0 to prevent division by zero

# Plot
plt.figure(figsize=(10, 6))

# Coulombic potential
U_coulomb = -Z * e2 / r
plt.plot(r, U_coulomb, 'k--', label='Coulomb: $-Ze^2/r$')

# Softened Coulomb potentials with various Rc
for Rc in Rc_values:
    U_erf = -Z * e2 * erf(r / Rc) / r
    plt.plot(r, U_erf, label=f'Soft Coulomb (Rc={Rc})')

# Formatting
plt.title("Comparison of Coulomb and Softened Coulomb Potentials")
plt.xlabel("r (a.u.)")
plt.ylabel("Potential Energy (a.u.)")
plt.ylim(-2, 0.2)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
